### [Q2.] What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0,2,50)
y = 3 + np.random.normal(0,2,len(x))
```

#### Note: Also add title, xlabel, and ylabel to the plot.

##### [Ans]

A scatter plot is a type of data visualization that uses dots to represent individual data points on a two-dimensional plane. Each dot's position corresponds to the values of two variables, one on the x-axis and the other on the y-axis. Scatter plots help visualize the relationship or correlation between two numerical variables and are useful for identifying patterns and trends in the data. They are commonly used in statistics, data science, and scientific research to explore data and assess the correlation between variables

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.title("SCATTER PLOT")
plt.show()
```

### [Q3.] Why is the subplot() function used? Draw four line plots using the subplot() function.

#### Use the following data:

```python
# FOR LINE 1:
x1 = np.array([0,1,2,3,4,5])
y1 = np.array([0,100,200,300,400,500])

# FOR LINE 2:
x2 = np.array([0,1,2,3,4,5])
y2 = np.array([50,20,40,20,60,70])

# FOR LINE 3:
x3 = np.array([0,1,2,3,4,5])
y3 = np.array([10,20,30,40,50,60])

# FOR LINE 4:
x4 = np.array([0,1,2,3,4,5])
y4 = np.array([200,350,250,550,450,150])
```

##### [Ans]

subplot( ) :- The subplot function in Matplotlib is used to create multiple plots within the same figure, allowing you to display several visualizations in a single window or canvas.

```python
import matplotlib.pyplot as plt

# CREATING ROWS AND COLUMNS FOR THE GRAPH :
fig, ((ax1,ax2) , (ax3,ax4)) = plt.subplots(2,2, figsize=(10,10))

# CREATING PLOT FOR LINE 1:
ax1.plot(x1,y1)
ax1.set_title("LINE 1")

# CREATING PLOT FOR LINE 2:
ax2.plot(x2,y2)
ax2.set_title("LINE 2")

# CREATING A PLOT FOR LINE 3:
ax3.plot(x3,y3)
ax3.set_title("LINE 3")

# CREATING A PLOT FOR LINE 4:
ax4.plot(x4,y4)
ax4.set_title("LINE 4")

# DISPLAYING ALL THE PLOTS:
plt.show()
```

### [Q4.] What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple" , "MIcrosoft" , "Google" , "AMD"])
profit = np.array([3000 , 8000 , 1000 , 10000])
```

##### [Ans]

A bar plot is a type of data visualization that represents categorical data using rectangular bars. Each bar's length corresponds to the value of a specific category, and the bars are usually spaced apart to show different categories distinctly.

Bar plots are used to compare discrete groups of data and visualize the distribution of values within each category.

```python
plt.bar(company , profit)
plt.xlabel("COMPANY")
plt.ylabel("PROFIT")
plt.title("BAR PLOT")
plt.show()
```

```python
plt.barh(company , profit)
plt.xlabel("PROFIT")
plt.ylabel("COMPANY")
plt.title("HORIZONTAL BAR PLOT")
plt.show()
```

### [Q5.] What is a box plot? Why is it used? Using the following data plot a box plot.

```python
box1 = np.random.normal(100,10,200)
box2 = np.random.normal(90,20,200)
```

##### [Ans]

A box plot, also known as a box-and-whisker plot, is a type of data visualization that provides a summary of the distribution of numerical data. It displays the median, quartiles, and potential outliers of a dataset in a compact and easy-to-understand format. The box plot consists of a rectangular "box" and "whiskers" that extend from the box.

It is commonly used for identifying outliers and helps to understanding the spread and skewness of the data.

```python
plt.boxplot([box1 , box2])
plt.show()
```
