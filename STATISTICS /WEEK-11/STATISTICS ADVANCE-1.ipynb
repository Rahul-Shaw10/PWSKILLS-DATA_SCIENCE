### [Q3.] Write a Python function to calculate the probability density function of a normal distribution with given mean and standard deviation at a given point.
##### [Ans]

```python
x = np.linspace(1,50,200)

def normal_distribution(x,mean,std):
    prob_density = (np.pi*std) * np.exp(-0.5*((x-mean)/std)**2)
    return prob_density

mean = np.mean(x)
std = np.std(x)

pdf = normal_distribution(x,mean,std)

plt.plot(x, pdf)
plt.xlabel("DATA POINTS")
plt.ylabel("PROBABILITY DENSITY")
plt.show()
```

### [Q4.] What are the properties of Binomial distribution? Give two examples of events where binomial distribution can be applied.
##### [Ans]

The properties of Binomial distribution are as follows:

1. The experiment consists of a fixed number of trials, say n.
2. The trials are independent and identical.
3. Each trial has only two possible outcomes, either a success or a failure.
4. The probability of success remains constant throughout all trials.
5. The random variable X, which represents the number of successes, follows a binomial distribution with parameters n and p.
6. Two examples of events where the binomial distribution can be applied are:

**Examples :** <br>
**Coin Tossing:** If we toss a coin n times and count the number of heads, then the number of heads follows a binomial distribution. <br>
**Quality Control:** In manufacturing, if we inspect n products and count the number of defective products, then the number of defective products follows a binomial distribution.

### [Q5.] Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4 and plot a histogram of the results using matplotlib.
##### [Ans]

```python
n = 10
p = 0.4
sample = np.random.binomial(n , p , 1000)

s = pd.Series(sample)
s.value_counts()
```

```python
s.value_counts(True).sort_index()
```

```python
plt.hist(s , bins=range(n+2) , align="left" , rwidth=0.8 , color="red")
plt.xticks(range(n+1))
plt.xlabel("NUMBER OF SUCCESSES")
plt.ylabel("FREQUENCY")
plt.title("BINOMIAL DISTRIBUTION WITH n=10 , p=0.4")
plt.show()
```

### [Q6.] Write a Python function to calculate the cumulative distribution function of a Poisson distribution with given mean at a given point.
##### [Ans]

```python
def possion_distribution(mean , k):
    cdf = 0
    for i in range(k+1):
        cdf += (math.exp(-mean) * mean**i) / math.factorial(i)
    return cdf

mean = 2.5
k = 3
cdf = possion_distribution(mean , k)
print(cdf)
```

### [Q7.] How Binomial distribution different from Poisson distribution?
##### [Ans]


| BINOMIAL DISTRIBUTION | POISSON DISTRIBUTION |
| --------------------- | -------------------- |
| Models the number of successes in a fixed number of trails. | Model the number of events in a fixed interval of time or space. |
| Trails are independent and have the same probability of success. | Events occur randomly and independently. |
| Used when the outcome is binary (success / failure). | Used when the events have a known average rate. |

### [Q8.] Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the sample mean and variance.
##### [Ans]

```python
sample = np.random.poisson(5 , 1000)
sample_mean = np.mean(sample)
sample_var = np.var(sample)

print("SAMPLE MEAN :",sample_mean)
print("SAMPLE VARIANCE :",sample_var)
```

**The sample and variance are likely to be close to their theoretical values 5.**

### [Q9.] How mean and variance are related in Binomial distribution and Poisson distribution?
##### [Ans]

In **Binomial distribution**, the mean and variance are related by the formula as follows:
- Mean $(x̄) = n * p$
- Variance $(σ) = n * p * (1-p)$

In **Poisson Distribution**, the mean and variance are related by the formula as follows:
- Mean $(x̄) = λ$
- Variance $(σ) = λ$

### [Q10.] In normal distribution with respect to mean position, where does the least frequent data appear?
##### [Ans]

In a normal distribution, the data is clustered near the mean becomes less frequent as it moves away from the mean. Therefore, the least frequent data appears in the tails of the distribution, which are the extreme ends.
