### [Q7.] You are a data analyst for a flight booking website, and you have been asked to analyze the Flight Price dataset to identify any trends in flight prices. What features would you analyze to identify these trends, and what visualizations would you use to present your findings to your team?
##### [ANS]

###### **OJECTIVE :**
The objective of the project is to identify the trends and visualizations.

###### **DATA USED :**
Data was used from UCI Machine Learning Repository which is freely available platform for data scientists and machine learning enthusiasts. We are using jupyter-notebook to run Flight Price Prediction task.

###### **DATA ANALYSIS :**

```python
import pandas as pd
df = pd.read_excel("flight_price.xlsx")
```

```python
df['Date_of_Journey'] = pd.to_datetime(df['Date_of_Journey'])
df['Month'] = df['Date_of_Journey'].dt.month
df['Day of week'] = df['Date_of_Journey'].dt.dayofweek
df['Year'] = df['Date_of_Journey'].dt.year
df['Day'] = df['Date_of_Journey'].dt.day
```

```python
avg_price_by_month = df.groupby('Month')['Price'].mean()
avg_price_by_dayofweek = df.groupby('Day of week')['Price'].mean()
avg_price_by_day = df.groupby('Day')['Price'].mean()
avg_price_by_year = df.groupby('Year')['Price'].mean()
```

```python
import matplotlib .pyplot as plt

plt.figure(figsize=(10,6))
plt.subplot(2,2,1)
plt.plot(avg_price_by_month , marker='o' , color='b')
plt.xlabel('MONTH')
plt.ylabel('AVERAGE PRICE')
plt.title('AVERAGE FLIGHT PRICES BY MONTH')
plt.grid(True)

plt.subplot(2,2,2)
plt.plot(avg_price_by_dayofweek , marker='o' , color='b')
plt.xlabel('DAY OF WEEK')
plt.ylabel('AVERAGE PRICE')
plt.title('AVERAGE FLIGHT PRICES BY DAY OF WEEK')
plt.grid(True)

plt.subplot(2,2,3)
plt.plot(avg_price_by_year , marker='o', color='b')
plt.xlabel('YEAR')
plt.ylabel('AVERAGE PRICE')
plt.title('AVERAGE FLIGHT PRICES BY YEAR')
plt.grid(True)

plt.subplot(2,2,4)
plt.plot(avg_price_by_day , marker='o' , color='b')
plt.xlabel('DAY')
plt.ylabel('AVERAGE PRICE')
plt.title('AVERAGE FLIGHT PRICES BY DAY')
plt.grid(True)

plt.tight_layout()
plt.show()
```

### [Q8.] You are a data scientist working for an airline company, and you have been asked to analyze the Flight Price dataset to identify the factors that affect flight prices. What features would you analyze to identify these factors, and how would you present your findings to the management team?
##### [ANS]

```python
import pandas as pd
import numpy as np
df = pd.read_excel("flight_price.xlsx")

df['Duration_hours'] = df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['Duration_minutes'] = df['Duration'].str.split(' ').str[1].str.split('m').str[0]

df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,np.nan:1})

df['Duration_minutes'] = df['Duration_minutes'].replace(np.nan , 0)
df['Duration_hours'] = df['Duration_hours'].replace('5m',5)

df['Duration_hours']=df['Duration_hours'].astype(int)
df['Duration_minutes']=df['Duration_minutes'].astype(int)

numerical_features = ['Duration_hours','Duration_minutes','Price','Total_Stops']
numerical_corr = df[numerical_features].corr()

plt.figure(figsize=(10,8))
sns.heatmap(numerical_corr, annot=True, fmt='.2f', linewidths=0.5)
plt.title("CORRELATION MATRIX OF NUMERICAL FEATURES")
plt.show()

categorical_features = ['Airline','Source','Destination','Additional_Info']
for feature in categorical_features:
    plt.figure(figsize=(10,6))
    sns.boxplot(x=feature, y='Price' , data=df)
    plt.title(f'BOXPLOT OF PRICE V/S {feature}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

### [Q15.] A company wants to launch a new app on the Google Playstore and has asked you to analyze the Google Playstore dataset to identify the most popular app categories. How would you approach this task, and what features would you analyze to make recommendations to the company?
##### [ANS]

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df1 = pd.read_csv('my.csv')

df1.drop_duplicates(inplace=True)

df1.dropna(subset=["Category","Installs"], inplace=True)

df1['Installs'] = df1['Installs'].str.replace('+','').str.replace(',','').astype(float)

total_installs_bycategory = df1.groupby('Category')['Installs'].sum().sort_values(ascending=False)
avg_rating_bycategory = df1.groupby('Category')['Rating'].mean().sort_values(ascending=False)

plt.figure(figsize=(12,8))

plt.subplot(1,2,1)
total_installs_bycategory.head(10).plot(kind='bar' , color='maroon' , edgecolor='black')
plt.xlabel("CATEGORY")
plt.ylabel("INSTALLS")
plt.title("TOP 10 CATEGORIES BY TOTAL INSTALLS")
plt.tight_layout()

plt.subplot(1,2,2)
avg_rating_bycategory.head(10).plot(kind='bar' , color='olive' , edgecolor='black')
plt.xlabel("CATEGORY")
plt.ylabel("RATINGS")
plt.title("TOP 10 CATEGORIES BY AVERAGE RATINGS")
plt.tight_layout()

plt.show()
```

### [Q16.] A mobile app development company wants to analyze the Google Playstore dataset to identify the most successful app developers. What features would you analyze to make recommendations to the company, and what data visualizations would you use to present your findings?

##### [ANS]

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


df1 = pd.read_csv("my.csv")

df1.drop_duplicates(inplace=True)

df1.dropna(subset=['App','Installs','Rating','Reviews','Size','Price'], inplace=True)

df1['Installs'] = df1['Installs'].str.replace('+','').str.replace(',','').astype(int)

total_apps_by_devop = df1.groupby('App')['App'].count().sort_values(ascending=False)

total_installs_by_devop = df1.groupby('App')['Installs'].sum().sort_values(ascending=False)

avg_rating_by_devop = df1.groupby('App')['Installs'].mean().sort_values(ascending=False)

total_review_by_devop = df1.groupby('App')['Reviews'].sum().sort_values(ascending=False)

df1['Size'] = df1['Size'].str.replace('M','').str.replace('k','').replace('Varies with device',np.nan).astype(float)
avg_size_by_devop = df1.groupby('App')['Size'].mean().sort_values(ascending=False)

df1['Price'] = df1['Price'].str.replace('$','').astype(float)
avg_price_by_devop = df1.groupby('App')['Price'].mean().sort_values(ascending=False)

plt.figure(figsize=(20,20))

plt.subplot(3,2,1)
total_apps_by_devop.head(10).plot(kind='bar', color='skyblue')
plt.xlabel("DEVELOPER")
plt.ylabel("TOTAL NUMBERS OF APPS")
plt.title("TOP 10 DEVELOPERS BY TOTAL NUMBER OF APPS")
plt.xticks(rotation=45)

plt.subplot(3,2,2)
total_installs_by_devop.head(10).plot(kind='bar', color='lightgreen')
plt.xlabel("DEVELOPER")
plt.ylabel("TOTAL INSTALLS")
plt.title("TOP 10 DEVELOPERS BY TOTAL INSTALLS")
plt.xticks(rotation=45)

plt.subplot(3,2,3)
avg_rating_by_devop.head(10).plot(kind='bar', color='lightcoral')
plt.xlabel("DEVELOPER")
plt.ylabel("AVERAGE RATING")
plt.title("TOP 10 DEVELOPERS BY AVERAGE RATING")
plt.xticks(rotation=45)

plt.subplot(3,2,4)
total_review_by_devop.head(10).plot(kind='bar', color='purple')
plt.xlabel("DEVELOPER")
plt.ylabel("TOTAL REVIEWS")
plt.title("TOP 10 DEVELOPERS BY TOTAL REVIEWS")
plt.xticks(rotation=45)

plt.subplot(3,2,5)
avg_size_by_devop.head(10).plot(kind='bar', color='orange')
plt.xlabel("DEVELOPER")
plt.ylabel("AVERAGE SIZES")
plt.title("TOP 10 DEVELOPERS BY AVERAGE SIZE")
plt.xticks(rotation=45)

plt.subplot(3,2,6)
avg_price_by_devop.head(10).plot(kind='bar', color='pink')
plt.xlabel("DEVELOPER")
plt.ylabel("AVERAGE PRICE")
plt.title("TOP 10 DEVELOPERS BY AVERAGE PRICE")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

### [Q17.] A marketing research firm wants to analyze the Google Playstore dataset to identify the best time to launch a new app. What features would you analyze to make recommendations to the company, and what data visualizations would you use to present your findings?

```python
import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.read_csv("my.csv")

df1['Last Updated'] = pd.to_datetime(df1['Last Updated'])

df1['Month'] = df1['Last Updated'].dt.month

monthly_update = df1.groupby('Month')['App'].count()

plt.figure(figsize=(10, 6))
monthly_update.plot(kind='line', marker='o', color='blue')
plt.title('MONTHLY TRENDS OF APP UPDATES')
plt.xlabel('MONTH')
plt.ylabel('NUMBER OF APP UPDATES')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```
