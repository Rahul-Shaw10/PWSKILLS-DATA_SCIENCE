### [Q2.] How did you handle missing data in the wine quality data set during the feature engineering process? Discuss the advantages and disadvantages of different imputation techniques.
##### [ANS]

Handling missing data is a crucial step in the feature engineering process, especially when working with datasets like the wine quality dataset. Different imputation techniques can be used to fill in missing values, each with its own advantages and disadvantages. Some common techniques include mean imputation, median imputation, mode imputation and using machine learning algorithms for imputation.

**1. MEAN IMPUTATION**:<br>
- *ADVANTAGES* : 
1. Simple and straightforward method.
2. Preserves the overall distribution of the data.

- *DISADVANTAGE* :
1. May introduce bias, especially if the data has outliers.
2. Does not account for correlations between features.

**2. MEDIAN IMPUTATION**:<br>
- *ADVANTAGES* :
1. Robust to outliers compared to mean imputation.
2. Preserves the central tendency of the data.

- *DISADVANTAGES* :
1. Ignores any relationships or patterns in the data.
2. May not be suitable for categorical variables.

**3. MODE IMPUTATION**:<br>
- *ADVANTAGES* :
1. Suitable for categorical variables.
2. Preserves the most frequent value in the dataset.

- *DIISADVANTAGES* :
1. Ignores relationships between variables.
2. May not be appropriate for continuous variables.

**4. USING MACHINE LEARNING ALGORITHMS**:<br>
- *ADVANTAGES* :
1. Utilizes relationships between variables to predict missing values.
2. Can handle both continuous and categorical variables.

- *DISADVANTAGES* :
1. Requires more computational resources.
2. Complexity increases with the number of features and missing data patterns.

### [Q5.] Load the wine quality data set and perform exploratory data analysis (EDA) to identify the distribution of each feature. Which feature(s) exhibit non-normality, and what transformations could be applied to these features to improve normality?
##### [ANS]

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("winequality.csv", delimiter=';')
df.head(2)
```

```python
plt.figure(figsize=(18,12))
df.hist(bins=10, color='green' ,edgecolor='black', grid=False)
plt.tight_layout()
plt.show()
```

```python
skewness = df.skew()
print("SKEWNESS OF EACH FEATURES :\n")
print(skewness)
```

```python
skewed_feature = skewness[abs(skewness) > 0.5].index.tolist()
print("\nFEATURES WITH HIGH SKEWNESS (> 0.5) :")
print(skewed_feature)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a single figure for all histograms
plt.figure(figsize=(12 ,8))

# Loop through the skewed features and plot histograms
for i, feature in enumerate(skewed_feature, start=1):
    plt.subplot(3, 3, i)
    sns.histplot(df[feature], kde=True, color='blue', bins=30)
    plt.title(f'Original Distribution of {feature}')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

### [Q6.] Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of features. What is the minimum number of principal components required to explain 90% of the variance in the data?
##### [ANS]

```python
df.head()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import numpy as np

X = df.drop('quality',axis=1)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA()
pca.fit(X_scaled)

explained_variance_ratio = pca.explained_variance_ratio_

cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

n_components_90_var = np.argmax(cumulative_variance_ratio >= 0.9) + 1

print("EXPLAINED VARIANCE RATIO FOR EACH PRINCIPAL :")
print(explained_variance_ratio)
print("\nCUMULATIVE EXPLAINED VARIANCE RATIO :")
print(cumulative_variance_ratio)
print("\nMINIMUN NUMBER OF PRINCIPAL COMPONENTS TO EXPLAIN 90% OF THE VARIANCE :", n_components_90_var)
```
