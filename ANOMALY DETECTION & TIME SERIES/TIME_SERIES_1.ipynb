
## Q1. What is a time series, and what are some common applications of time series analysis?
#####[Ans]
A **time series** is a sequence of data points recorded at successive time intervals. It is used to analyze and predict patterns over time.  
**Common Applications**:
1. **Finance**: Stock price prediction, risk analysis.
2. **Retail**: Sales forecasting, inventory management.
3. **Healthcare**: Patient monitoring, disease spread analysis.
4. **Energy**: Power consumption forecasting.
5. **Weather**: Temperature and rainfall prediction.

---

## Q2. What are some common time series patterns, and how can they be identified and interpreted?
#####[Ans]
1. **Trend**: Long-term increase or decrease in data values. Identified through visualization or smoothing techniques.
2. **Seasonality**: Regular repeating patterns at specific intervals (e.g., monthly sales). Detected using seasonal decomposition.
3. **Cyclic Patterns**: Fluctuations occurring over irregular intervals, linked to external factors (e.g., economic cycles).
4. **Irregular Variations**: Random noise or anomalies not explained by trends or seasonality.

---

## Q3. How can time series data be preprocessed before applying analysis techniques?
#####[Ans]
1. **Handling Missing Values**: Use imputation techniques like forward fill, backward fill, or interpolation.
2. **Smoothing**: Apply moving averages or exponential smoothing to reduce noise.
3. **Normalization/Scaling**: Scale data to ensure consistent measurement units.
4. **Stationarization**: Remove trends and seasonality using differencing or decomposition.
5. **Outlier Detection**: Identify and handle outliers using statistical methods or visualization.

## Q4. How can time series forecasting be used in business decision-making, and what are some common challenges and limitations?
#####[Ans]
**Uses in Business**:
1. **Inventory Management**: Predicting demand to optimize stock levels.
2. **Budget Planning**: Forecasting revenue and expenses.
3. **Marketing**: Timing promotional campaigns based on predicted trends.

**Challenges and Limitations**:
1. **Data Quality**: Missing or noisy data affects accuracy.
2. **Model Limitations**: Cannot account for sudden, unpredicted events.
3. **Stationarity**: Non-stationary data may require complex transformations.

## Q5. What is ARIMA modelling, and how can it be used to forecast time series data?
#####[Ans]
**ARIMA (AutoRegressive Integrated Moving Average)** is a statistical model for time series forecasting that combines:
1. **AR (AutoRegressive)**: Models the relationship with previous time points.
2. **I (Integrated)**: Differencing to make the series stationary.
3. **MA (Moving Average)**: Models the relationship with past forecast errors.

ARIMA is useful for short-term forecasting and works well with stationary time series.

---

## Q6. How do Autocorrelation Function (ACF) and Partial Autocorrelation Function (PACF) plots help in identifying the order of ARIMA models?
#####[Ans]
- **ACF**: Shows correlation of a time series with its lagged values. Used to determine the order of the **MA** component.
- **PACF**: Shows correlation of a time series with its lagged values, removing influence of intermediate lags. Used to determine the order of the **AR** component.

---

## Q7. What are the assumptions of ARIMA models, and how can they be tested for in practice?
#####[Ans]
**Assumptions**:
1. The series is stationary (constant mean and variance over time).
2. Errors are normally distributed and independent.

**Testing**:
1. **Stationarity**: Perform Augmented Dickey-Fuller (ADF) or KPSS test.
2. **Error Distribution**: Use Q-Q plots or Shapiro-Wilk test.

## Q8. Suppose you have monthly sales data for a retail store for the past three years. Which type of time series model would you recommend for forecasting future sales, and why?
#####[Ans]
A **SARIMA (Seasonal ARIMA)** model is recommended because:
1. The data likely exhibits **seasonality** (monthly trends).
2. SARIMA handles both non-seasonal and seasonal components effectively.
3. Provides accurate short-to-medium term forecasts.

## Q9. What are some of the limitations of time series analysis? Provide an example of a scenario where the limitations of time series analysis may be particularly relevant.
#####[Ans]
**Limitations**:
1. Cannot account for external factors (e.g., sudden market changes).
2. Requires high-quality, continuous data.
3. Assumes past patterns will continue into the future.

**Example**: Predicting stock prices during a financial crisis is challenging because market conditions deviate significantly from historical patterns.

---

## Q10. Explain the difference between a stationary and non-stationary time series. How does the stationarity of a time series affect the choice of forecasting model?
#####[Ans]
- **Stationary Series**: Mean, variance, and autocorrelation are constant over time. Suitable for ARIMA and similar models.
- **Non-Stationary Series**: Mean or variance changes over time. Requires transformation (e.g., differencing).

**Impact on Models**:
1. Stationary series enable simpler models like ARIMA.
2. Non-stationary series need pre-processing to ensure accurate forecasts.
